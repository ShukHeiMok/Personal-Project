```python
# Data analysis and Manipulation
import plotly.graph_objs as go
import plotly.io as pio
import plotly.express as px
import pandas as pd

# Data Visualization
import matplotlib.pyplot as plt

# Importing Plotly
import plotly.offline as py
py.init_notebook_mode(connected=True)

# Initializing Plotly
pio.renderers.default = 'colab'
```

```python
# Importing Dataset1
dataset1 = pd.read_csv("covid.csv")
dataset1.head(n=10)  # returns first 5 rows
```

```python
dataset2=pd.read_csv("covid_grouped.csv")
dataset2.head(n=10)
```

```python
dataset3=pd.read_csv("coviddeath.csv")
dataset3.head(n=10)
```

```python
# Returns tuple of shape (Rows, columns), size
print(dataset1.shape, dataset1.size)

print(dataset2.shape, dataset2.size)

print(dataset3.shape, dataset3.size)
```

```python
# Drop NewCases, NewDeaths, NewRecovered rows from dataset1 Column
dataset1.drop(['NewCases', 'NewDeaths', 'NewRecovered'], 
              axis=1, inplace=True)

# Select random set of values from dataset1
dataset1.sample(5)
```

```python
# Import create_table Figure Factory

from plotly.figure_factory import create_table

colorscale = [[0, '#4d004c'], [.5, '#f2e5ff'], [1, '#ffffff']]
table = create_table(dataset1.head(15), colorscale=colorscale)
py.iplot(table)
```

```python
px.bar(dataset1.head(15), x = 'Country/Region', y = 'TotalCases',
       color = 'TotalDeaths', height = 500,
       hover_data = ['Country/Region', 'Continent', 'TotalDeaths'])
```

```python
px.bar(dataset1.head(15), x = 'Country/Region', y = 'TotalTests',
       color = 'TotalCases', height = 500, hover_data = ['Country/Region', 'Continent'])
```

```python
px.bar(dataset1.head(15), x = 'TotalTests', y = 'Continent',
       color = 'TotalTests',orientation ='h',  height = 500,
       hover_data = ['Country/Region', 'Continent'])
```

```python
px.scatter(dataset1, x='Continent',y='TotalCases', 
           hover_data=['Country/Region', 'Continent'], 
           color='TotalCases', size='TotalCases', size_max=80)
```
